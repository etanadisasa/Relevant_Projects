## PART ONE
#### To start working, we will create three dataframes.

```python
import pandas as pd
employees = {
        'Employees': ['David', 'Hailé', 'Addis', 'Alem', 'Solomn'],
        'Dep_Id': ['93', '02', '02', '93', '90'], 
        'Branch_Id': ['01', '02', '03', '05', '07']}
df_emp = pd.DataFrame(employees, columns = ['Employees', 'Dep_Id', 'Branch_Id'])
df_emp
```

```python
departments = {
        'DepName': ['Accounts', 'CustomerRelation', 'EmployeeHealth', 'HR', 'Leadership', 'Sales', 'Utilities'],
        'Dep_Id': ['01', '02', '25', '55', '85', '90', '93']}
df_dep = pd.DataFrame(departments, columns = ['DepName', 'Dep_Id'])
df_dep
```

```python
branches = {
        'BranchName': ['Arada', 'Bole', 'Megenagna', 'Piassa', 'Saris', 'Warit', 'Yeka'],
        'Branch_Id': ['01', '02', '03', '04', '05', '06', '07']}
df_bra = pd.DataFrame(branches, columns = ['BranchName', 'Branch_Id'])
df_bra
```

### Merge with Left Join

```python
pd.merge(df_emp,df_dep, on='Dep_Id', how='left')
```

### Merge with Right Join

```python
pd.merge(df_emp,df_dep, on='Dep_Id', how='right')
```

### Merge with Inner Join

```python
pd.merge(df_emp,df_bra, on='Branch_Id', how='inner')
```

### Merge with Outer Join

```python
pd.merge(df_emp,df_bra, on='Branch_Id', how='outer')
```

### Correct Employee Dep and Branch Data Table
#### 1) Employee Left Join Department Left Join Branch

```python

df_empdep = pd.merge(df_emp,df_dep, on='Dep_Id', how='left')
pd.merge(df_empdep,df_bra, on='Branch_Id', how='left')
```

#### 2) Employee Left Join Branch Inner Join Department

```python
df_empbra = pd.merge(df_emp,df_bra, on='Branch_Id', how='left')
pd.merge(df_empbra,df_dep, on='Dep_Id', how='inner')
```

```python
AirData = pd.read_csv('AirQualityUCI.csv', delimiter=";")
#AirData.head()
AirData.tail()
```

```python
AirData.shape # This dataset has 9471 rows and 17 columns. Rows after row 9357 all reguster NaN value. 
```

```python
AirData.info()
```

#### It appears that in this dataset, there are several null values. Observations after row 9356 across all the variables have NaN values. Furthermore, two unnamed columns exist with no observations. Two ways we can deal with this problem is. 
#### 1) Deleting the two columns and the several rows using loc() function

```python
AirData_Edit1 = AirData.loc[0:9356,'Date':'AH']
AirData_Edit1.tail()
```

#### 2) Deleting the two columns using loc[] and drop() functions

```python
AirData_Edit2 = AirData.drop(['Unnamed: 15', 'Unnamed: 16'], axis=1)
AirData_Edit2 = AirData_Edit2.loc[0:9356]
AirData_Edit2.tail()
```

#### These two methods can get rid of both rows with NaN values and also columns with NaN values.
